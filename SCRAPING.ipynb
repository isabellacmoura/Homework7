```python
from bs4 import BeautifulSoup
import requests

response = requests.get("https://www.lemonde.fr")
doc = BeautifulSoup(response.text)
```

```python

items = doc.find_all(class_='article')
len(items)
```

```python
items
```

```python

for item in items:
    print('----')
    print("Title:", item.select_one('p.article__title').get_text(strip=True) if item.select_one('p.article__title') else "")

    if item.name == "a" and item.has_attr('href'):
        print(item['href'])
    else:
        print("No link found") 
    
    print("Subhead:", item.select_one('.article__subhead, .article__desc, .subhead').get_text(strip=True) if item.select_one('.article__subhead, .article__desc, .subhead') else "No subhead found")
    print("Type:", item.select_one('.article__kicker').get_text(strip=True) if item.select_one('.article__kicker') else "")
    print("Author:", item.select_one('.article__author-name, .article__byline').get_text(strip=True) if item.select_one('.article__author-name, .article__byline') else " No type found")
   
    print("Image URL:", item.select_one('img').get('data-lazy') or item.select_one('img').get('src') or item.select_one('img').get('data-srcset') or item.select_one('img').get('srcset') if item.select_one('img') else "No image found")

    print("Premium:", "Premium" if item.select_one('.icon__premium, .article__premium') else "")
```

```python
import pandas as pd
rows1 = []

for item in items:
    row1 = {}
    if item.select_one('p.article__title'):
        row1['title'] = item.select_one('p.article__title').get_text(strip=True)
    elif item.select_one('img') and item.select_one('img').has_attr('alt'):
        row1['title'] = item.select_one('img')['alt'].strip()
    else:
        row1['title'] = ""
    row1['url'] = item['href'] if item.has_attr('href') else (item.select_one('a')['href'] if item.select_one('a') and item.select_one('a').has_attr('href') else "")
    img = item.select_one('img')
    row1['image'] = (
        img.get('data-lazy') or img.get('src') or img.get('data-srcset') or img.get('srcset')
        if img else "No image found"
    )
    row1['subhead'] = (
        item.select_one('.article__subhead, .article__desc, .subhead').get_text(strip=True)
        if item.select_one('.article__subhead, .article__desc, .subhead') else "No subhead found"
    )
    row1['type'] = item.select_one('.article__kicker').get_text(strip=True) if item.select_one('.article__kicker') else ""
    row1['author'] = (
        item.select_one('.article__author-name, .article__byline').get_text(strip=True)
        if item.select_one('.article__author-name, .article__byline') else "No author found"
    )
    row1['premium'] = "premium" if item.select_one('.icon__premium, .article__premium') else ""
    rows1.append(row1)

len(rows1)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://travel-id-documents.az.gov/authorized-third-party-driver-license-locations'
headers = {'User-Agent': 'Mozilla/5.0'}

response = requests.get(url, headers=headers)
response.raise_for_status()
```

```python
doca = BeautifulSoup(response.text, 'html.parser')
rows = doca.select('table tbody tr')

data = []

for row in rows:
    cols = row.find_all('td')
    if len(cols) >= 4:
        company_tag = cols[0].find('a')
        if company_tag:
            company_name = company_tag.get_text(strip=True)
            company_link = company_tag['href']
        else:
            company_name = cols[0].get_text(strip=True)
            company_link = ''

        address = cols[1].get_text(strip=True)
        phone = cols[2].get_text(strip=True)


        data.append({
            'Company': company_name,
            'Company Link': company_link,
            'Address': address,
            'Phone': phone,
        })


```

```python
df = pd.DataFrame(data)
df.head()


```

```python
df.to_csv('Authorized_Third_Party_Driver.csv', index=False)
```

```python
from bs4 import BeautifulSoup
import requests

response = requests.get("https://www.tnwb.uscourts.gov/Search/Search.aspx?zoom_sort=0&zoom_xml=0&zoom_query=cars&zoom_per_page=150&zoom_and=1&zoom_cat%5B%5D=-1")
car = BeautifulSoup(response.text)
```

```python
items = car.select('.result_block, .result_altblock')
len(items) 
```

```python
import os

for item in items:
    print("---------")
    print(item.select_one('.result_title').text.split('[')[0].strip())
    print(item.find('a')['href'])
    print(item.select_one('.category').text)  
    
    if item.select_one('.infoline').text:
        info_text = item.select_one('.infoline').text
        separator = info_text.split(' - ')
        
        if len(separator) >= 3:
            print("terms matched:", separator[0].replace("Terms matched:", "").strip())
            size_number = separator[1].strip().split('k')[0]
            print(f"size: {size_number}k")
            pdf_url = separator[2].replace("URL:", "").strip()
            print("pdf:", pdf_url)

            if pdf_url.endswith('.pdf'):
                file_name = os.path.basename(pdf_url)
                try:
                    response = requests.get(pdf_url)
                    with open(file_name, "wb") as f:
                        f.write(response.content)
                except Exception as e:
                    print(f"Error.")
    
    
```

```python
import pandas as pd
rows = []

for item in items:
    row = {}
    
    row['case'] = item.select_one('.result_title').text.split('[')[0].strip()
    row['url'] =   item.find('a')['href']
    row['category'] =item.select_one('.category').text

    if item.select_one('.infoline').text:
        info_text = item.select_one('.infoline').text
        separator = info_text.split(' - ')
        
        if len(separator) >= 3:
            row['terms matched'] = separator[0].replace("Terms matched:", "").strip()
            row['size'] =  separator[1].strip().split('k')[0]
            row['pdf_url'] = separator[2].replace("URL:", "").strip()


    rows.append(row)
len(rows)
```

```python
df = pd.json_normalize(rows)
df.head()
```
